## 1. Treating Null Values

```python
round(df.isna().sum()/len(df)*100, 2)
```

```python
def null_values(data, null_pencentage):
    null_values = len(df) * null_pencentage
    columns_to_drop = []
    for col in data.columns:
        if data[col].isna().sum() > null_values:
            columns_to_drop.append(col)
    return columns_to_drop
            
```

```python
drop_columns = null_values(df, 0.4)
drop_columns
```

```python
data = df.drop(columns=drop_columns)
```

```python
data.isna().sum().sum()
```

```python
print(null_values(data, 0.15))
print('median: ', data['LotFrontage'].median())
print('skewness: ', data['LotFrontage'].skew())
```

```python
data['LotFrontage'].fillna(data['LotFrontage'].median(), inplace = True)
```

## 4. Cleaning Operations 

```python
def cleaning_operations(data):
    return(
        data
        .assign(
            MSZoning = data.MSZoning.str.replace(' (all)', ''),
            
            LotFrontage = data.LotFrontage.astype('int'),
            
            YrSold = pd.to_datetime(data['YrSold']).dt.year,
            YearBuilt = pd.to_datetime(data['YearBuilt']).dt.year,
            GarageYrBlt = pd.to_datetime(data['GarageYrBlt']).dt.year,
            YearRemodAdd = pd.to_datetime(data['YearRemodAdd']).dt.year,
            
            
            
        )
    )
```
